VARIABLES

```python
dbutils.widgets.text("catalog_name", "catalog_smartdata")
dbutils.widgets.text("raw_schema", "raw")
dbutils.widgets.text("bronze_schema", "bronze")
dbutils.widgets.text("silver_schema", "silver")
dbutils.widgets.text("gold_schema", "gold")
dbutils.widgets.text("storage_account", "adlsjrdata")
```

```python
catalog_name = dbutils.widgets.get("catalog_name")
raw_schema = dbutils.widgets.get("raw_schema")
bronze_schema = dbutils.widgets.get("bronze_schema")
silver_schema = dbutils.widgets.get("silver_schema")
gold_schema = dbutils.widgets.get("gold_schema")
storage_account = dbutils.widgets.get("storage_account")
```

```python
data_uri = f"abfss://raw@{storage_account}.dfs.core.windows.net/"
```

CREANDO LOS SCHEMAS

```python
spark.sql(f"USE CATALOG {catalog_name}")

schemas = [raw_schema, bronze_schema, silver_schema, gold_schema]

for schema in schemas:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{schema}")

print("Schemas creados correctamente")
```
